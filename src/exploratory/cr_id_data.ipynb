```python
import pandas as pd

officer_list_path = '/Users/jonathanjuarez/invisible-institute/data/update_2024/final-profiles.csv'
officer_df = pd.read_csv(officer_list_path)
investigators_path = '/Users/jonathanjuarez/invisible-institute/data/update_2024/investigators.csv'
investigator_df = pd.read_csv(investigators_path)
```

```python
print('officer_df columns:',officer_df.columns)
print('investigator_df columns:',investigator_df.columns)
```

```python
# Create full name fields for officers and investigators
officer_df['full_name'] = officer_df['first_name'].str.title() + ' ' + officer_df['last_name'].str.title()
investigator_df['full_name'] = investigator_df['first_name'].str.title() + ' ' + investigator_df['last_name'].str.title()

# Combine the names into lists
full_names = officer_df['full_name'].tolist() + investigator_df['full_name'].tolist()
first_names = officer_df['first_name'].str.title().tolist() + investigator_df['first_name'].str.title().tolist()
last_names = officer_df['last_name'].str.title().tolist() + investigator_df['last_name'].str.title().tolist()

# Deduplicate the lists
full_names = list(set(full_names))
first_names = list(set(first_names))
last_names = list(set(last_names))
```

```python
print(full_names)
```

```python
# Remove NaN values from the list
cleaned_full_names = [name for name in full_names if pd.notna(name)]
```

```python
# Export the cleaned list to a .txt file
file_path = '/Users/jonathanjuarez/Documents/Advanced ML/NLP-Police-Complaints/data/cleaned_full_names.txt'
with open(file_path, 'w') as file:
    for name in cleaned_full_names:
        file.write(name + '\n')

print(f"Exported {len(cleaned_full_names)} names to {file_path}")
```

```python
street_names_path = "/Users/jonathanjuarez/Documents/Advanced ML/NLP-Police-Complaints/Chicago_Street_Names_20240302.csv"
streets_df = pd.read_csv(street_names_path)
streets_df.columns
```

```python
names_df = pd.DataFrame(cleaned_full_names, columns=['Entity'])
names_df['Label'] = 'PERSON'

# Prepare the street names from streets_df
streets_df['Entity'] = streets_df['Full Street Name'].astype(str).str.title()  # Ensure consistent case formatting
streets_df = streets_df.drop_duplicates(subset=['Entity'])  # Drop any duplicates
streets_df['Label'] = 'LOC'

# Combine the dataframes
combined_df = pd.concat([names_df, streets_df[['Entity', 'Label']]], ignore_index=True)

# Save to CSV
combined_df.to_csv('combined_entities.csv', index=False)

print(f"Exported combined entities to combined_entities.csv")
```
